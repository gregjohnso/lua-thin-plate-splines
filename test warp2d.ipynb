```lua
require './warp2d'
require 'image'

im = image.load('lena.png')
im2 = image.rotate(im, math.pi)

itorch.image(im)
for i = 1,5 do
    pts_anchor, pts_def = warp2d.gen_warp_pts(im:size(), 3, 12)
    im_warp, warpfield = warp2d.warp(im, pts_anchor, pts_def)
    im2_warp = image.warp(im2, warpfield)
    itorch.image(torch.cat(torch.sum(im_warp,1), torch.sum(im2_warp,1), 3))
end
```
